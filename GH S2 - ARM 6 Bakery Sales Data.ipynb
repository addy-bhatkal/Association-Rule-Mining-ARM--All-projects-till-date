```python
df = pd.read_csv('ARM - Bakery.csv')
```

```python
df['Items'] = df['Items'].str.strip()
```

```python
df['Items'].apply(type).unique()
```

```python
df['Daypart'].value_counts()
```

```python
basket = df[df['Daypart']=='Afternoon'].groupby(['TransactionNo','Items'])['DateTime'].sum().unstack().reset_index().fillna(0).set_index('TransactionNo')
basket
```

```python
def encode(x):
    if x== 0:
        return 0
    if x!= 0:
        return 1
```

```python
basket = basket.applymap(encode)
basket
```

```python
freq = apriori(basket, min_support=0.005, use_colnames=True)
freq
```

```python
rules = association_rules(freq, metric='lift', min_threshold=1)

rules = rules.sort_values(by='lift', ascending = False)
rules
```
